```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cdk_all = pd.read_csv("Single_protein/cdk_all.csv")
```

```python
response_cols = []
for col in cdk_all.filter(like="avg_pIC50", axis=1).columns.values:
    response_cols.append(col)
response_cols
```

```python
pIC50 = cdk_all[response_cols].dropna()
combined_data = pd.concat([pIC50[col] for col in pIC50.columns])
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(combined_data, bins=15)
plt.title('Combined Distribution of avg_pIC50_cdk Values')
plt.xlabel('avg_pIC50_cdk Values')
plt.ylabel('Frequency')
plt.show()
```
